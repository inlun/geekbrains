```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
# УРОК 2 ЗАДАНИЕ 1.2) Напишите код на Python, реализующий расчет длины 
# вектора, заданного его координатами. (в программе)

x = [5, 2, 3]
print(math.sqrt(sum([el ** 2 for el in x])))
```

```
# УРОК 2 ЗАДАНИЕ 3.1) Напишите код на Python, реализующий построение
# графиков: окружности 

# в данном случае окружность с центром в координатах (3,5) и радиусом 10

x = []
y1 = []
y2 = []
for i in range(-7000, 13001):
    x_ = i/1000
    x.append(x_)
    y1.append(math.sqrt(10 ** 2 - (x_ - 3) ** 2) + 5)
    y2.append(-math.sqrt(10 ** 2 - (x_ - 3) ** 2) + 5)    
plt.plot(x,y1)
plt.plot(x,y2)
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
```

```
# УРОК 2 ЗАДАНИЕ 3.2) Напишите код на Python, реализующий построение 
# графиков: эллипса 

# в данном случае эллипс с фокусом +/- sqrt(5) на оси абсцисс

x = []
y1 = []
y2 = []
for i in range(-3000, 3001):
    x_ = i/1000
    x.append(x_)
    y1.append(math.sqrt((1 - x_ ** 2 / 3 ** 2) * 2 ** 2))
    y2.append(-math.sqrt((1 - x_ ** 2 / 3 ** 2) * 2 ** 2))    
plt.plot(x,y1)
plt.plot(x,y2)
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
```

```
# УРОК 2 ЗАДАНИЕ 3.2) Напишите код на Python, реализующий построение 
# графиков: гиперболы 

x1 = []
x2 = []
y = []
for i in range(-10000, 10000):
    y_ = i/1000
    y.append(y_)
    x1.append(3 * math.sqrt(1 + y_ ** 2 / 2 ** 2))
    x2.append(3 * (-1) * math.sqrt(1 + y_ ** 2 / 2 ** 2))
plt.plot(x1,y)
plt.plot(x2,y)
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
```

```
# УРОК 3 ЗАДАНИЕ 4.1) Решите систему уравнений:
#
# y = x2 – 1
# exp(x) + x∙(1 – y) = 1

# графический метод
x = np.linspace(-5, 5, 10000)
plt.plot(x, x ** 2 - 1)
plt.plot(x, 1 - (1 - np.exp(x)) / x)

# численный метод
from scipy.optimize import fsolve
def equations(p):
    x, y = p
    return (y - x ** 2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (-2, 5))
x2, y2 =  fsolve(equations, (2, 5))
x3, y3 =  fsolve(equations, (6, 25))
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
